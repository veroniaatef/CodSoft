```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
from sklearn.utils import resample
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
Credit_Card=pd.read_csv('creditcard.csv')
```

```python
Credit_Card.head()
```

```python
# Display the last few rows of the dataset
Credit_Card.tail()
```

```python
# Display the total number of elements in the dataset
Credit_Card.size
```

```python
# Convert column names to lowercase
Credit_Card.columns=Credit_Card.columns.str.lower()
```

```python
Credit_Card.columns
```

```python
# Remove duplicate rows based on specific columns
Credit_Card.drop_duplicates(subset=['time', 'v1', 'v2', 'v3', 'v4', 'v5', 'v6', 'v7', 'v8', 'v9', 'v10',
       'v11', 'v12', 'v13', 'v14', 'v15', 'v16', 'v17', 'v18', 'v19', 'v20',
       'v21', 'v22', 'v23', 'v24', 'v25', 'v26', 'v27', 'v28', 'amount',
       'class'],keep='first',inplace=True)
```

```python
Credit_Card.info()
```

```python
# Plot a countplot to visualize the distribution of classes
plt.figure(figsize=(15,5))
sns.countplot(data=Credit_Card,x='class')
```

```python
# Display the count of each class
Credit_Card['class'].value_counts().to_frame()
```

```python
# Separate majority and minority classes
fraud_majority=Credit_Card[Credit_Card['class']==0]
fraud_minority=Credit_Card[Credit_Card['class']==1]
```

```python
# Downsample the majority class to balance the dataset
fraud_majority_downsample=resample(fraud_majority,n_samples=473,replace=False,random_state=42)
```

```python
# Concatenate the downsampled majority class with the minority class
fraud_dataset=pd.concat([fraud_majority_downsample,fraud_minority])
```

```python
# Display the count of each class in the balanced dataset
fraud_dataset['class'].value_counts().to_frame()
```

```python
# Plot a countplot to visualize the distribution of classes in the balanced dataset
plt.figure(figsize=(15,5))
sns.countplot(data=fraud_dataset,x='class')
```

```python
# Calculate the correlation between features and the target variable (class)
dataset_corr_class=fraud_dataset.corr()['class'].to_frame()
```

```python
# Plot a barplot to visualize the correlation between features and the target variable
plt.figure(figsize=(15,10))
sns.barplot(data=dataset_corr_class,y=dataset_corr_class.index,x='class')
```

```python
# Separate features (x) and target variable (y)
x=fraud_dataset.drop(['class'],axis=1)
y=fraud_dataset['class']
```

```python
# Split the dataset into training and testing sets then display their shapes
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
# Initialize and train a logistic regression model
lr=LogisticRegression(max_iter=500)
lr.fit(x_train,y_train)
```
